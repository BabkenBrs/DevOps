```python
#Достанем подкатегории
def extract_subcategories(cat_link):
    sub_links = []
    page = requests.get(cat_link)
    soup = bs4.BeautifulSoup(page.content, 'html.parser')
    for link in soup.find_all('a', class_='sc-dQoVA dsEnQi products-slider__header'):
        href = link.get("href")
        if 'cat/c' in href:
            sub_links.append('http://perekrestok.ru' + href)
    return sub_links
```

```python
def cal_info(page):
    info = page.find_all('div', class_ = 'product-calories-item__value')
    titles = page.find_all('div', class_ = 'product-calories-item__title')
    carb = None
    fats = None
    proteins = None
    calor = None
    for i in range(len(titles)):
        if titles[i].get_text() == 'Жиры':
            fats = info[i].get_text().split()[0]
            continue
        if titles[i].get_text() == 'Углеводы':
            carb = info[i].get_text().split()[0]
            continue
        if titles[i].get_text() == 'Белки':
            proteins = info[i].get_text().split()[0]
            continue
        if titles[i].get_text() == 'Ккал':
            calor = info[i].get_text().split()[0]
            continue
        
    return carb, fats, proteins, calor


def extract_info(item_link):
    link = requests.get(item_link)
    page = bs4.BeautifulSoup(link.content, 'html.parser')
    img_link = page.find_all('img', itemprop='image')[0].get('src')
    item_name = page.find_all('h1', itemprop = 'name')[0].get_text()
    carb, fats, proteins, calories = cal_info(page)
    vendor_name = item_link.split('-')[-1]
    item = {'vendor_name' : vendor_name,
            'img' : img_link, 
            'name' : item_name,
            'carbohydrates' : carb,
            'fats' : fats,
            'proteins' : proteins,
            'calories' : calories}
    return item
```

```python
import time
import pandas as pd

def creating_DB(list_of_cats):
    DataBase = ({
    'vendor_name' : [],
    'img' : [],
    'name' : [],
    'carbohydrates' : [],
    'fats' : [],
    'proteins' : [],
    'calories' : []
    })

    df = pd.DataFrame(DataBase)
    
    for cat in list_of_cats[:8]:
        list_of_subcats = extract_subcategories(cat)
        for subcat in list_of_subcats[:5]:
            item_links = extract_item_links_from_subcat(subcat)
            for item in item_links[:5]:
                print(item)
                item_info = extract_info(item)
                time.sleep(0.5)
                df = df._append(item_info, ignore_index = True)
    return df
```
